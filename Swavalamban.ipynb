```python
import datetime
df_phase=df.groupby('phase')
df_phase_startup=df_phase.get_group('Start-up')
df_phase_startup=df_phase_startup[['phase','enterprise startup/scaleup Date']]
#conversion of date to date time (because if we have only date then we cannot use strftime fn)
df_phase_startup['date'] = pd.to_datetime(df_phase_startup['enterprise startup/scaleup Date'], yearfirst=True)
df_phase_startup['Year'] = df_phase_startup['date'].dt.strftime('%Y')
x=pd.DataFrame(df_phase_startup['Year'])
x.columns=['year']
xz=x.groupby('year').size().reset_index()
xz
```

```python
#Count of candidates in start up phase after 2017
plt.rcParams['font.family'] = 'Arial Unicode MS'
style.use("ggplot")
x=xz['year']
y=xz[0]
plt.figure(figsize=(14,7))
plt.xlabel("year",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Count of candidates in start up phase after 2017",fontsize=18)
plt.bar(x,y,color="red",alpha=0.9,width=0.5)
plt.show()
```

```python
#count of candidates in scale up phase on the basis of year
df_phase_scaleup=df_phase.get_group('Scale-up')
df_phase_scaleup=df_phase_scaleup[['phase','enterprise startup/scaleup Date']]
x=pd.DatetimeIndex(df_phase_scaleup['enterprise startup/scaleup Date']).year
x=pd.DataFrame(np.array(x))
x.columns=['year']
xz=x.groupby('year').size().reset_index()
xz
```

```python
plt.rcParams['font.family'] = 'Arial Unicode MS'
style.use("ggplot")
x=xz['year']
y=xz[0]
plt.figure(figsize=(14,7))
plt.xlabel("year",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Count of candidates in scale up phase from 2020-2023",fontsize=18)
plt.bar(x,y,color="green",alpha=0.9,width=0.5)
plt.show()
```
